# **AIML Project**


**In this AIML project we are using AI algorithms to predict the Delaney Solubility**

What is **Delaney Solubility ?**

Answer : **Delaney solubility**, also known as ESOL (Estimated SOLubility), is a method for **estimating the aqueous solubility of a compound** directly from its molecular structure. The model was derived from a set of 2874 measured solubilities using **linear regression** against nine molecular properties.

This data is often used in the field of data science, particularly in machine learning, to predict **compound solubility**. For example, it’s used in Kaggle competitions and in projects on GitHub for solubility prediction. This allows researchers and data scientists to build models that can predict the solubility of new, untested compounds, which is very useful in fields like drug discovery.

#Importing Libraries

Importing the required libraries to our python environment for training our AI model which predicts the logS value.

```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
```

#Data Preparation



```python
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
print(df)
```

```python
y = df['logS']
x = df.drop('logS',axis=1) #axis 1 takes data as column format
# splitting x and y values
```

##Separation of Data

in this step we are separating our data into two sets one is for training and other one is for the testing phase.

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=100)
```

#Model Building

in this we are going to build a model for the identification of the LogS

##Linear Regression

training the model to predict the data, in this we are using sklearn AI alogrithm for linear regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train,y_train)
```

###Applying the model to make prediction on given data

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
print(y_lr_train_pred)
```

```python
print(y_lr_test_pred)
```

####Evaluating the model

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2 = r2_score(y_train,y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2 = r2_score(y_test,y_lr_test_pred)
```

```python
lr_results = pd.DataFrame([['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]],columns=['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2'])
lr_results
```

## Random Forest

training the model to predict the data, in this we are using sklearn AI alogrithm for RandomForest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(x_train,y_train)
```

###Applying the model to make prediction on given data

```python
y_rf_train_pred = lr.predict(x_train)
y_rf_test_pred = lr.predict(x_test)
```

```python
print(y_rf_train_pred)
```

```python
print(y_rf_test_pred)
```

#### Evaluating the model

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train,y_rf_train_pred)
rf_train_r2 = r2_score(y_train,y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2 = r2_score(y_test,y_rf_test_pred)
```

```python
rf_results = pd.DataFrame([['Random Forest Regressor', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]],columns=['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2'])
rf_results
```

## Model Comparison

```python
df_models = pd.concat([lr_results, rf_results], axis=0).reset_index(drop=True)
df_models
```

#Data Visualization

```python
plt.figure(figsize=(5,5))
plt.scatter(x=y_train,y=y_lr_train_pred,c='#416D19',alpha=0.3)

z = ny.polyfit(y_train,y_lr_train_pred,1)
p = ny.poly1d(z)

plt.plot(y_train,p(y_train),'red')
plt.ylabel('Predicted LogS of LR')
plt.xlabel('Experimental LogS')


plt.figure(figsize=(5,5))
plt.scatter(x=y_train,y=y_rf_train_pred,alpha=0.3)

z = ny.polyfit(y_train,y_rf_train_pred,1)
p = ny.poly1d(z)

plt.plot(y_train,p(y_train),'red')
plt.ylabel('Predicted LogS of RF')
plt.xlabel('Experimental LogS')
```
